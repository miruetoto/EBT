```python
!pip uninstall PyEBT -y
!pip install git+https://github.com/miruetoto/PyEBT.git
import numpy as np
import matplotlib.pyplot as plt 
import PyEBT
#from PyPkg.PyEBT.core import ebt
plt.rcParams['text.usetex'] = True
```

```python
fs = 1000
t = np.arange(-2000, 2000 + 1) / fs
v1 = np.sin(2 * np.pi * t)
v2 = np.cos(20 * np.pi * t)
f = v1 * v2
```

```python
out1 = PyEBT.ebt(t,f,tau=60)
i1 = out1['knot'][0]
i2 = out1['knot'][20]
i3 = out1['knot'][40]
out2 = PyEBT.ebt(t,f,tau=80)
i4 = out2['knot'][0]
i5 = out2['knot'][20]
i6 = out2['knot'][40]
```

```python
# 플롯 설정
fig, axs = plt.subplots(3, 2, figsize=(10, 10), dpi=150)

# 1. Original Signal
axs[0, 0].plot(t, f, color='k')
axs[0, 0].set_title("Original Signal")
axs[0, 0].set_xlim(-0.25, 0.25)
axs[0, 0].set_xlabel("")
axs[0, 0].set_ylabel("")

# 2. Single Touch of Dessin
axs[0, 1].plot(t, f, color='k')
axs[0, 1].scatter(t[i1], out1['band'][i1,0], color='red', s=20)
axs[0, 1].plot(t[i1], out1['band'][i1,0], color='red', alpha=0.2, linewidth=1)
axs[0, 1].set_title("Single Touch of Dessin")
axs[0, 1].set_xlim(-0.22, 0.22)

# 3. Three Touches
axs[1, 0].plot(t, f, color='k')
axs[1, 0].scatter(t[i1], out1['band'][i1,0], color='red', s=20, label="Band 1")
axs[1, 0].plot(t[i1], out1['band'][i1,0], color='red', alpha=0.2, linewidth=1)
axs[1, 0].scatter(t[i2], out1['band'][i2,20], color='blue', s=20, label="Band 20")
axs[1, 0].plot(t[i2], out1['band'][i2,20], color='blue', alpha=0.2, linewidth=1)
axs[1, 0].scatter(t[i3], out1['band'][i3,40], color='green', s=20, label="Band 40")
axs[1, 0].plot(t[i3], out1['band'][i3,40], color='green', alpha=0.2, linewidth=1)
axs[1, 0].set_title("Three Touches")
axs[1, 0].legend()
axs[1, 0].set_xlim(-0.25, 0.25)

# 4. All Possible Touches
axs[1, 1].plot(t, f, color='k')
for i in range(out1['band'].shape[1]):
    axs[1, 1].plot(t, out1['band'][:,i], color='gray', linewidth=0.05)
axs[1, 1].set_title("All Possible Touches")
axs[1, 1].scatter(t[i1], out1['band'][i1,0], color='red', s=20, label="Band 1")
axs[1, 1].plot(t[i1], out1['band'][i1,0], color='red', alpha=0.2, linewidth=1)
axs[1, 1].scatter(t[i2], out1['band'][i2,20], color='blue', s=20, label="Band 20")
axs[1, 1].plot(t[i2], out1['band'][i2,20], color='blue', alpha=0.2, linewidth=1)
axs[1, 1].scatter(t[i3], out1['band'][i3,40], color='green', s=20, label="Band 40")
axs[1, 1].plot(t[i3], out1['band'][i3,40], color='green', alpha=0.2, linewidth=1)
axs[1, 1].set_xlim(-0.25, 0.25)

# 5. Three Touches (With Larger Interval)
axs[2, 0].plot(t, f, color='k')
axs[2, 0].scatter(t[i4], out2['band'][i4,0], color='red', s=20, label="Band 1")
axs[2, 0].plot(t[i4], out2['band'][i4,0], color='red', alpha=0.2, linewidth=1)
axs[2, 0].scatter(t[i5], out2['band'][i5,20], color='blue', s=20, label="Band 20")
axs[2, 0].plot(t[i5], out2['band'][i5,20], color='blue', alpha=0.2, linewidth=1)
axs[2, 0].scatter(t[i6], out2['band'][i6,40], color='green', s=20, label="Band 40")
axs[2, 0].plot(t[i6], out2['band'][i6,40], color='green', alpha=0.2, linewidth=1)
axs[2, 0].set_title("Three Touches (With Larger Interval)")
axs[2, 0].legend()
axs[2, 0].set_xlim(-0.25, 0.25)

# 6. All Possible Touches (With Larger Interval)
axs[2, 1].plot(t, f, color='k')
axs[2, 1].scatter(t[i4], out2['band'][i4,0], color='red', s=20, label="Band 1")
axs[2, 1].plot(t[i4], out2['band'][i4,0], color='red', alpha=0.2, linewidth=1)
axs[2, 1].scatter(t[i5], out2['band'][i5,20], color='blue', s=20, label="Band 20")
axs[2, 1].plot(t[i5], out2['band'][i5,20], color='blue', alpha=0.2, linewidth=1)
axs[2, 1].scatter(t[i6], out2['band'][i6,40], color='green', s=20, label="Band 40")
axs[2, 1].plot(t[i6], out2['band'][i6,40], color='green', alpha=0.2, linewidth=1)
for i in range(out2['band'].shape[1]):
    axs[2, 1].plot(t, out2['band'][:, i], color='gray', linewidth=0.05)
axs[2, 1].set_title("All Possible Touches (With Larger Interval)")
axs[2, 1].set_xlim(-0.22, 0.22)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```
